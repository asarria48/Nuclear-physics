#Funcion Uniforme

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Parámetros de la distribución uniforme
a = 2
b = 8
c = 1/(b-a)  # altura constante
mu = (a + b) / 2  # media
sigma = (b - a) / np.sqrt(12)  # desviación estándar

# Valor z para el 68% del área
z = 0.34 * (b - a)

# Crear la figura
plt.figure(figsize=(10, 6))

# Dibujar la distribución uniforme
x = np.linspace(a-1, b+1, 1000)
y = np.where((x >= a) & (x <= b), c, 0)
plt.plot(x, y, 'b-', linewidth=2, label=f'Distribución uniforme u(x)')

# Sombreado del 68% del área
x_fill = np.linspace(mu - z, mu + z, 100)
y_fill = np.where((x_fill >= a) & (x_fill <= b), c, 0)
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.5, label='68% del área')

# Líneas verticales para los límites a y b
plt.axvline(a, color='red', linestyle='--', linewidth=1.5, label='Límites')
plt.axvline(b, color='red', linestyle='--', linewidth=1.5)

# Media y desviación estándar
plt.axvline(mu, color='green', linestyle='--', linewidth=1.5, label='μ')
plt.axvline(mu - sigma, color='orange', linestyle='--', linewidth=1.5, label='μ ± σ')
plt.axvline(mu + sigma, color='orange', linestyle='--', linewidth=1.5)

# Rango del 68%
plt.axvline(mu - z, color='magenta', linestyle=':', linewidth=2, label='μ ± z')
plt.axvline(mu + z, color='magenta', linestyle=':', linewidth=2)

# Gráfico
plt.xlabel('x')
plt.ylabel('u(x)')
plt.title('Distribución Uniforme')
plt.ylim(0, c * 1.5)
plt.xlim(a-1, b+1)
plt.xticks([])
plt.yticks([])
plt.text(a, -0.01, f'a ', ha='center', color='red')
plt.text(b, -0.01, f'b ', ha='center', color='red')
plt.text(mu*1.02, c*1.02, f'μ', ha='center', color='green')
plt.text(mu - sigma, -0.01, f'μ-σ ', ha='center', color='orange')
plt.text(mu + sigma, -0.01, f'μ+σ ', ha='center', color='orange')
plt.text(mu - z, -0.01, f'μ-z ', ha='center', color='magenta')
plt.text(mu + z, -0.01, f'μ+z ', ha='center', color='magenta')
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("RESUMEN DE VALORES CALCULADOS:")
print(f"a = {a}")
print(f"b = {b}")
print(f"c = 1/(b-a) = {c:.4f}")
print(f"μ = (a+b)/2 = {mu}")
print(f"σ = (b-a)/√12 = {sigma:.4f}")
print(f"Área entre μ-σ y μ+σ = {2*sigma*c:.4f} = {2*sigma*c*100:.2f}%")
print(f"z para 68% del área = 0.34*(b-a) = {z:.4f}")
print(f"Intervalo para 68%: [μ-z, μ+z] = [{mu-z:.2f}, {mu+z:.2f}]")
print(f"Área sombreada = {2*z*c:.4f} = {2*z*c*100:.2f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

#DADO PIRAMIDAL

valores = np.array([1, 2, 3, 4, 5, 6])
probabilidades = np.array([1/6, 1/4, 1/6, 1/6, 1/12, 1/6])

# Media
media = np.sum(valores * probabilidades)

# Varianza
varianza = np.sum((valores - media)**2 * probabilidades)
desviacion = np.sqrt(varianza)

# Asimetría
asimetria = np.sum(((valores - media)/desviacion)**3 * probabilidades)

print("Resultados del dado piramidal:")
print(f"Valor medio (μ): {media:.4f}")
print(f"Varianza: {varianza:.4f}")
print(f"Desviación estándar (σ): {desviacion:.4f}")
print(f"Asimetría (τ1): {asimetria:.4f}")

plt.bar(valores, probabilidades, color="#1f77b4", edgecolor="black", alpha=0.7)
plt.axvline(media, color="red", linestyle="--", label=f"Media = {media:.2f}")
plt.axvline(media - desviacion, color="green", linestyle="--", label=f"μ - σ = {media-desviacion:.2f}")
plt.axvline(media + desviacion, color="green", linestyle="--", label=f"μ + σ = {media+desviacion:.2f}")
plt.axvline(media + asimetria, color="yellow", linestyle="--", label=f"μ + τ = {media+asimetria:.2f}")


plt.title("Distribución de probabilidad del dado piramidal")
plt.xlabel("Valor del dado")
plt.ylabel("Probabilidad")
plt.xticks(valores)
plt.legend()
plt.grid(axis="y", linestyle=":")
plt.show()
```

```python
# Importo las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Leo los datos (debo previamente subir el archivo amplificador.dat al colab, o desde la caja de código previa a esta)
datos = pd.read_csv('amplificador.dat', sep='\s+')

# Extraigo los valores de voltaje (en este caso puedo quitar el ['mV'] porque tengo solo una columna, pero es útil para otros .dat con más de una columna)
voltajes = datos['mV'].values

# Muestro información del .dat referentes a los voltajes (núm. datos, 5 valores a los extremos)
print(f"Número de mediciones: {len(voltajes)}")
print(f"Primeros 5 valores: {voltajes[:5]}")
print(f"Últimos 5 valores: {voltajes[-5:]}")
print()

media = np.sum(voltajes) / len(voltajes)                    # Saco la media como la conocía en la ecuación (7)
varianza = np.sum((voltajes - media)**2) / len(voltajes)    # Saco la varianza
desviacion = np.sqrt(varianza)                              # Saco la desviación estándar, raíz cuadrada de la varianza, como la conocía en la ecuación (8)


# Función para calcular mu y sigma
def stats_from_hist(data, deltaV):                                      # Recibe los datos y el ancho
    vmin, vmax = data.min(), data.max()
    bins = np.arange(vmin, vmax + deltaV, deltaV)                       # Defino los intervalos
    freq, edges = np.histogram(data, bins=bins)                         # Cuento las frecuencias para armar el histograma
    mids = 0.5 * (edges[:-1] + edges[1:])                               # Acomodo el centro de cada bin
    mu = np.sum(mids * freq) / np.sum(freq)                             # Calculo la media
    sigma = np.sqrt(np.sum(((mids - mu) ** 2) * freq) / np.sum(freq))   # Calculo la desviación
    return mu, sigma

# Aplico la función para los voltajes propuestos del problema, saco mu y sigma
mu_25, sigma_25 = stats_from_hist(voltajes, 2.5)
mu_10, sigma_10 = stats_from_hist(voltajes, 10)

# Para graficar, primero configuro
vmin, vmax = voltajes.min(), voltajes.max()                             # Extremos
rango = vmax - vmin                                                     # Rango total
extra = 0.05 * rango                                                    # Para un mejor visual
limites = (vmin - extra, vmax + extra)                                  # Ampliación con el extra

# Definición de los bins para ambos histogramas (de cada ancho propuesto)
bins1 = np.arange(limites[0], limites[1], 2.5)
bins2 = np.arange(limites[0], limites[1], 10)

fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)            # Creo dos subgráficos para poner los dos histogramas en un mismo plot

# --- Histograma ΔV = 2.5 mV ---
freq1, edges1, patches1 = axes[0].hist(voltajes, bins=bins1, edgecolor='black',   # Acomodo la función histograma de Python de acuerdo a mis preferencias (para la anchura de 2.5 mV)
                                      alpha=0.7, color='#1f77b4')

# Añado las líneas de media y desviación estándar
axes[0].axvline(mu_25, color='red', linestyle='--', linewidth=2,
               label=f'μ = {mu_25:.1f} mV')
axes[0].axvline(mu_25 - sigma_25, color='orange', linestyle=':',
               label=f'μ ± σ = [{mu_25-sigma_25:.1f}, {mu_25+sigma_25:.1f}]')
axes[0].axvline(mu_25 + sigma_25, color='orange', linestyle=':')

# Rótulos, título, y grid para el gráfico
axes[0].set_xlabel("Voltaje (mV)")
axes[0].set_ylabel("Frecuencia")
axes[0].set_title("Histograma - ΔV = 2.5 mV")
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].set_xlim(limites)


# Y lo mismo para la otra anchura propuesta
# --- Histograma ΔV = 10 mV ---
freq2, edges2, patches2 = axes[1].hist(voltajes, bins=bins2, edgecolor='black',
                                      alpha=0.7, color='#1f77b4')

axes[1].axvline(mu_10, color='red', linestyle='--', linewidth=2,
               label=f'μ = {mu_10:.1f} mV')
axes[1].axvline(mu_10 - sigma_10, color='orange', linestyle=':',
               label=f'μ ± σ = [{mu_10-sigma_10:.1f}, {mu_10+sigma_10:.1f}]')
axes[1].axvline(mu_10 + sigma_10, color='orange', linestyle=':')

axes[1].set_xlabel("Voltaje (mV)")
axes[1].set_ylabel("Frecuencia")
axes[1].set_title("Histograma - ΔV = 10 mV")
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].set_xlim(limites)

plt.tight_layout()
plt.show()                      # Muestro la figura aquí mismo en el cuaderno
plt.savefig("p3histos.pdf")     # Y también la guardo en .pdf


# Imprimo los valores que preguntaba el problema para media y desviación, usando las tres formas
print('Media original =', round(media, 4), 'mV')
print('Desviación estándar original =', round(desviacion, 4), 'mV')
print('Media ΔV=2.5 =', round(mu_25, 4), 'mV')
print('Desviación estándar ΔV=2.5 =', round(sigma_25, 4), 'mV')
print('Media ΔV=10 =', round(mu_10, 4), 'mV')
print('Desviación estándar ΔV=10 =', round(sigma_10, 4), 'mV')
print('Rango de datos =', round(vmin, 1), '-', round(vmax, 1), 'mV')
print('Límites de gráfico =', [round(limites[0], 1), round(limites[1], 1)])
print()

# Hago una tabla comparativa de los resultados de los tres casos
tabla = pd.DataFrame({
    "Caso": ["Valores originales", "Histograma ΔV = 2.5 mV", "Histograma ΔV = 10 mV"],
    "μ (mV)": [round(media, 2), round(mu_25, 2), round(mu_10, 2)],
    "σ (mV)": [round(desviacion, 2), round(sigma_25, 2), round(sigma_10, 2)]
})


# Imprimo la tabla para verla aquí mismo en el cuaderno
print("Tabla comparativa de resultados:")
print(tabla)
print()

# Imprimo la diferencia entre los valores de la media y la desviación originales con los nuevos, para ver qué tanto cambian
print("Análisis adicional:")
print(f"• Diferencia media ΔV=2.5 vs original: {abs(media - mu_25):.4f} mV")
print(f"• Diferencia media ΔV=10 vs original: {abs(media - mu_10):.4f} mV")
print(f"• Diferencia σ ΔV=2.5 vs original: {abs(desviacion - sigma_25):.4f} mV")
print(f"• Diferencia σ ΔV=10 vs original: {abs(desviacion - sigma_10):.4f} mV")
```

#Geiger

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Datos
d = np.array([147, 152, 153, 171, 146, 168, 145, 133, 168, 171])
N = len(d)
sigma_i = np.sqrt(d)

# Promedio
mu = np.mean(d)

# Desviación estándar de los datos
sigma = np.sqrt(np.mean((d - mu)**2))

# Crear tabla con pandas
tabla = pd.DataFrame({
    "i": np.arange(1, N+1),
    "d (cuentas)": d,
    "σ (cuentas)": np.round(sigma_i, 2)
})

print("\Tabla")
print(tabla.to_string(index=False))
# Incertidumbre del promedio (propagación)
sigma_mu = np.sqrt(np.sum(sigma_i**2) / N**2)

print("Valor promedio μ =", round(mu, 2))
print("Desviación estándar σ =", round(sigma, 2))
print("Incertidumbre propagada σμ =", round(sigma_mu, 2))

# Gráfica con barras de error
x = np.arange(1, N+1)
plt.errorbar(x, d, yerr=sigma_i, fmt='o', capsize=5, label="Datos con incertidumbre")
plt.axhline(mu, color="red", linestyle="--", label=f"Promedio μ = {mu:.2f}")
plt.axhline(mu + sigma_mu, color="green", linestyle=":", label=f"μ ± σμ")
plt.axhline(mu - sigma_mu, color="green", linestyle=":")
plt.xlabel("Número de medición (i)")
plt.ylabel("Conteo (d)")
plt.legend()
plt.grid(True, linestyle=":")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Datos de la Tabla 4
d = np.array([147, 152, 153, 171, 146, 168, 145, 133, 168, 171])
N = len(d)
sigma_i = np.sqrt(d)

# Valor promedio usando ecuación (8)
mu = np.sum(d) / N

# Incertidumbre usando estadística de Poisson
sigma_poisson = np.sqrt(mu)

# Incertidumbre del promedio por propagación
sigma_mu = np.sqrt(np.sum(sigma_i**2)) / N

# Crear tabla con pandas
tabla = pd.DataFrame({
    "i": np.arange(1, N+1),
    "d (cuentas)": d,
    "σ (cuentas)": np.round(sigma_i, 2)
})

print("Tabla")
print(tabla.to_string(index=False))

print(f"\nValor promedio μ = {mu:.2f} cuentas")
print(f"Incertidumbre (Poisson) σ = √μ = {sigma_poisson:.2f} cuentas")
print(f"Incertidumbre del promedio (propagación) σ_μ = {sigma_mu:.2f} cuentas")

# Gráfica paso a paso
plt.figure(figsize=(10, 6))

# 4.2: Puntos con barras de error
plt.errorbar(np.arange(1, N+1), d, yerr=sigma_i, fmt='o', capsize=5,
             label="Mediciones ± σ", color='blue', alpha=0.7)

# 4.3: Línea del valor promedio
plt.axhline(mu, color="red", linestyle="--", linewidth=2,
            label=f"Promedio μ = {mu:.2f}")

# 4.6: Bandas de incertidumbre del promedio
plt.axhline(mu + sigma_mu, color="green", linestyle=":", linewidth=1.5,
            label=f"μ ± σ_μ = {mu:.2f} ± {sigma_mu:.2f}")
plt.axhline(mu - sigma_mu, color="green", linestyle=":", linewidth=1.5)
plt.fill_between([0.5, N+0.5], mu - sigma_mu, mu + sigma_mu,
                 color='green', alpha=0.1)

plt.xlabel("Medición (i)")
plt.ylabel("Cuentas (d)")
plt.title("Mediciones con detector Geiger-Müller")
plt.legend()
plt.grid(True, linestyle=":", alpha=0.7)
plt.xlim(0.5, N+0.5)
plt.tight_layout()
plt.show()
```

#Ajuste + Chi cuadrado

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de la Sec. 2.2.1 de AnalisisDatos.pdf.
x = np.array([6.5, 9.3, 14.6])        # t [s]
y = np.array([3.2, 9.7, 11.3])        # x [cm]
sigma = np.array([1.2, 2.1, 3.6])     # cm

# la relación va a ser x(t) = a0 + a1t

# Caso A y B
def two_points(t1, x1, t2, x2):
    a1 = (x2 - x1) / (t2 - t1)        # valor de la pendiente
    a0 = x1 - a1 * t1                 # valor de la intersección
    return a0, a1

a0_A, a1_A = two_points(6.5, 3.2, 9.3, 9.7)     # primeros dos puntos
a0_B, a1_B = two_points(6.5, 3.2, 14.6, 11.3)   # primer y último punto

# Caso C
weights = 1 / sigma**2                          # pesos
S = np.sum(weights)                             # de aquí en adelante, todas las sumas ponderadas que están en la forma analítica
S_x = np.sum(x * weights)
S_y = np.sum(y * weights)
S_xx = np.sum(x**2 * weights)
S_xy = np.sum(x * y * weights)

delta = S * S_xx - S_x**2                         # determinante del sistema de ecuaciones
a1_C = (S * S_xy - S_x * S_y) / delta
a0_C = (S_xx * S_y - S_x * S_xy) / delta

# Caso D --> suponiendo que para t = 14.6 s, σ(x = 11.3 s) = 30.0 cm.
sigma_D = sigma.copy()
sigma_D[2] = 30.0                   # solo para el último dato (t = 14.6 s)
weights_D = 1 / sigma_D**2          # todo otra vez
S_D = np.sum(weights_D)
S_xD = np.sum(x * weights_D)
S_yD = np.sum(y * weights_D)
S_xxD = np.sum(x**2 * weights_D)
S_xyD = np.sum(x * y * weights_D)

delta_D = S_D * S_xxD - S_xD**2
a1_D = (S_D * S_xyD - S_xD * S_yD) / delta_D
a0_D = (S_xxD * S_yD - S_xD * S_xyD) / delta_D

# chi^2
def chi2(a0, a1, t, y, sigma):
    y_pred = a0 + a1 * t
    return np.sum(((y - y_pred) / sigma)**2)

chi2_A = chi2(a0_A, a1_A, x, y, sigma)
chi2_B = chi2(a0_B, a1_B, x, y, sigma)
chi2_C = chi2(a0_C, a1_C, x, y, sigma)
chi2_D = chi2(a0_D, a1_D, x, y, sigma_D)


x_fit = np.linspace(5, 16, 100)
yfit_A = a0_A + a1_A * x_fit
yfit_B = a0_B + a1_B * x_fit
yfit_C = a0_C + a1_C * x_fit
yfit_D = a0_D + a1_D * x_fit

y_A = a0_A + a1_A * x_fit
y_B = a0_B + a1_B * x_fit
y_C = a0_C + a1_C * x_fit
y_D = a0_D + a1_D * x_fit

fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)


# Panel A
axes[0].errorbar(x, y, yerr=sigma, fmt='o', capsize=3, label='Datos')
labelA = fr"$x(t) = {a0_A:.2f} + {a1_A:.2f}\,t$ $\chi^2 = {chi2_A:.2f}$"
axes[0].plot(x_fit, yfit_A, 'r-', label=labelA)
axes[0].set_title("Caso A")
axes[0].set_xlabel("t (s)")
axes[0].set_ylabel("x (cm)")
axes[0].legend(fontsize=9, loc='upper left')
axes[0].grid(True, linestyle='--', alpha=0.6)

# Panel B
axes[1].errorbar(x, y, yerr=sigma, fmt='o', capsize=3, label='Datos')
labelB = fr"$x(t) = {a0_B:.2f} + {a1_B:.2f}\,t$ $\chi^2 = {chi2_B:.2f}$"
axes[1].plot(x_fit, yfit_B, 'r-', label=labelB)
axes[1].set_title("Caso B")
axes[1].set_xlabel("t (s)")
axes[1].legend(fontsize=9, loc='upper left')
axes[1].grid(True, linestyle='--', alpha=0.6)

# Panel C
axes[2].errorbar(x, y, yerr=sigma, fmt='o', capsize=3, label='Datos')
labelC = fr"$x(t) = {a0_C:.2f} + {a1_C:.2f}\,t$ $\chi^2 = {chi2_C:.2f}$"
axes[2].plot(x_fit, yfit_C, 'r-', label=labelC)
axes[2].set_title("Caso C")
axes[2].set_xlabel("t (s)")
axes[2].legend(fontsize=9, loc='upper left')
axes[2].grid(True, linestyle='--', alpha=0.6)



# Panel D
# axes[3].errorbar(x, y, yerr=sigma_D, fmt='o', capsize=3, label='Datos')
# labelD = fr"$x(t) = {a0_D:.2f} + {a1_D:.2f}\,t$ $\chi^2 = {chi2_D:.2f}$"
# axes[3].plot(x_fit, yfit_D, 'r-', label=labelD)
# axes[3].set_title("Caso D")
# axes[3].set_xlabel("t (s)")
# axes[3].legend(fontsize=9, loc='upper left')
# axes[3].grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

#Klein-Nishina

```python
import numpy as np
import matplotlib.pyplot as plt

m_e_c2 = 511.0                                                         # 0.511 MeV = 511 keV: masa en reposo del electrón
E = np.array([10.0, 100.0, 511.0, 1333.0])                             # distintos valores de energía del fotón incidente
labels = [f"{int(E)}" for E in E]


def klein_nishina(theta, E):
    epsilon = E / m_e_c2                                               # energía reducida del fotón
    cos_t = np.cos(theta)
    denom = (1.0 + epsilon * (1-cos_t))
    pref = 1.0 / denom**2
    return 0.5 * pref * (1.0 + cos_t**2 + (epsilon**2 * (1-cos_t)**2) / denom)


theta_full = np.linspace(0, 2*np.pi, 721)

colors = ['C0', 'C1', 'C2', 'C3']
curves_full = [klein_nishina(theta_full, E) for E in E]


fig = plt.figure(figsize=(12,5))


axp = fig.add_subplot(1,2,1, projection='polar')                        # gráfica polar
for c, lab, col in zip(curves_full, labels, colors):
    axp.plot(theta_full, c, label=lab, color=col)

axp.set_thetagrids(np.arange(0, 360, 30))
for label, angle in zip(axp.get_xticklabels(), np.arange(0, 360, 30)):
    if angle > 180:
        label.set_visible(False)
axp.set_theta_zero_location("E")   # 0° hacia la derecha
axp.set_theta_direction(1)         # sentido antihorario


axl = fig.add_subplot(1,2,2)
theta_deg = np.linspace(0, 180, 361)
for E, lab, col in zip(E, labels, colors):
    axl.plot(theta_deg, klein_nishina(np.radians(theta_deg), E), color=col)    # gráfico lineal

axl.set_xticks(np.arange(0,181,30))
axl.set_xlabel("θ (°)")
axl.set_ylabel("dσ/dΩ ($r_e^2$)")
axl.set_title("Klein–Nishina (lineal)")
plt.minorticks_on()
plt.grid(True, which='major', linestyle='-', linewidth=0.7)
plt.grid(True, which='minor', linestyle='--', linewidth=0.4)

axp.legend(loc='upper right',title="$E_1$ [keV]", bbox_to_anchor=(1.35, 1.1))
axp.set_title("Klein–Nishina (polar)", pad=30)
plt.tight_layout()

plt.tight_layout()
plt.show()
```
